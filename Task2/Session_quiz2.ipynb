## Question 2.12

```python
from math import atan2, degrees
# Enter values
x=float(input("Give a real value for x: "))
y=float(input("Give a real value for y: "))
s_r=atan2(x, y) #radians
s_d=degrees(s_r) #degrees
print(s_d)

# Defining Quadrants

I=0 < s_d < 90
II= 90 < s_d < 180
III = 180 < s_d < 270
IV= 270 < s_d < 360

# Conditions

if I:
    print("Angle belongs to Quadrant I")
elif II:
    print("Angle belongs to Quadrant II")
elif III:
    print("Angle belongs to Quadrant III")
else:
    print("Angle belongs to Quadrant IV")
```

## Question 2.19

```python
import numpy as np

# mean and standard deviation
mu = 0
sigma = 1.4

s = np.random.normal(mu, sigma, 100000)

#Check mean and standard deviation
print(abs(mu - np.mean(s)))
print(abs(sigma - np.std(s, ddof=1)), "\n")

print("the standard deviation is ", np.std(s))
```
