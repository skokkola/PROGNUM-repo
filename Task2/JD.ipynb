```python
Y = int(input("Enter Year: "))
M = int(input("Enter Month: "))
D = float(input("Enter Day: "))
JD = 367*Y -7*(Y+(M+9)/12)/4 - 3*((Y+(M-9)/7)/100 + 1)/4 + (275*M)/9 + D + 1721029-0.5
print("Julian Date is ", JD)
```

```python
Y1 = int(input("Enter First Year: "))
M1 = int(input("Enter First Month: "))
D1 = float(input("Enter First Day: "))
JD1 = 367*Y1 -7*(Y1+(M1+9)/12)/4 - 3*((Y1+(M1-9)/7)/100 + 1)/4 + (275*M1)/9 + D1 + 1721029-0.5

Y2 = int(input("Enter Second Year: "))
M2 = int(input("Enter Second Month: "))
D2 = float(input("Enter Second Day: "))
JD2 = 367*Y2 -7*(Y2+(M2+9)/12)/4 - 3*((Y2+(M2-9)/7)/100 + 1)/4 + (275*M2)/9 + D2 + 1721029-0.5

print("The difference in days is ", abs(JD1-JD2), "days")
```
