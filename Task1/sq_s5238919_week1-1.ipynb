Wildcards are abbreviations for filenames or sets of filenames

1. command ls *[0-9]*.[co] shows all files with number between 0-9 and that ends in .c OR .o but not .co.

2. The brackets [ ] focus the command on within the brackets, e.g. [0-9] refers to numbers ranging from 0-9.  It chooses only one digit of the options, e.g. [co] gives c OR o.

3. the command ls *[0-9]*.co lists only files that have numbers ranging from 0-9 AND end in .co
